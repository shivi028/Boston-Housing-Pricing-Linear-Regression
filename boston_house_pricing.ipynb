```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv(r'C:\Users\shivi\OneDrive\Desktop\housing.csv')
dataset
```

```python
#independent featuresX-
x = dataset.drop(columns = ['MEDV'])
#dependent features
y = dataset['MEDV']
```

```python
print(x.head())   #independent feature
print(y.head())   #dependent feature MEDV
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=42)
```

```python
X_train
```

```python
# using linear regression
# STANDARDIZING DATASET
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
```

```python
X_train = scalar.fit_transform(X_train)
```

```python
X_test = scalar.transform(X_test)
```

```python
# # cross validation
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
```

```python
regression = LinearRegression()
regression.fit(X_train, y_train)
```

```python
mse = cross_val_score(regression, X_train, y_train, scoring='neg_mean_squared_error', cv=5)
```

```python
np.mean(mse)
```

```python
# predection
reg_predict = regression.predict(X_test)
```

```python
reg_predict
```

```python
import seaborn as sns
sns.displot(reg_predict-y_test, kind='kde')
```

```python
from sklearn.metrics import r2_score
```

```python
score = r2_score(reg_predict, y_test)
```

```python
score
```

```python
##Ridge Redression AlgoRithm   (remove overfitted data)
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV   #cross validation
```

```python
ridge_reg = Ridge()
```

```python
ridge_reg
```

```python
parameters = {"alpha": [1,2,3,4,10,20,30,40,50,60,70]}
ridgecv = GridSearchCV(ridge_reg, parameters, scoring='neg_mean_squared_error', cv=5)
ridgecv.fit(X_train, y_train)
```

```python
print(ridgecv.best_params_)
```

```python
print(ridgecv.best_score_)
```

```python
ridge_pred = ridgecv.predict(X_test)
```

```python
import seaborn as sns
sns.displot(ridge_pred-y_test, kind='kde')
```

```python
score = r2_score(ridge_pred, y_test)
```

```python
score
```

```python
from sklearn.linear_model import Lasso     #absolute value of slop
```

```python
lasso = Lasso()
```

```python
parameters = {"alpha": [1,2,3,4,10,20,30,40,50,60,70]}
lassocv = GridSearchCV(lasso, parameters, scoring='neg_mean_squared_error', cv=5)
lassocv.fit(X_train, y_train)
```
